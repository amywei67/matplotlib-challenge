```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
clinical_data_complete = pd.merge(clinical_data, mouse_drug_data, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
clinical_data_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_volumes_mean = clinical_data_complete.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_volumes_mean = pd.DataFrame(tumor_volumes_mean)

# Flatten the column headings
tumor_volumes_mean = tumor_volumes_mean.reset_index()

# Preview DataFrame
tumor_volumes_mean
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_volumes_sem = clinical_data_complete.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_volumes_sem = pd.DataFrame(tumor_volumes_sem)

# Preview DataFrame
tumor_volumes_sem.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_volumes_mean = tumor_volumes_mean.reset_index()
tumor_volumes_pivot_mean = tumor_volumes_mean.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

tumor_volumes_sem = tumor_volumes_sem.reset_index()
tumor_volumes_pivot_sem = tumor_volumes_sem.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
tumor_volumes_pivot_mean.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_volumes_pivot_mean.index, tumor_volumes_pivot_mean["Capomulin"], yerr=tumor_volumes_pivot_sem["Capomulin"], color="b", marker="o", markersize=5, linestyle="dashed", linewidth=0.70)
plt.errorbar(tumor_volumes_pivot_mean.index, tumor_volumes_pivot_mean["Infubinol"], yerr=tumor_volumes_pivot_sem["Infubinol"], color="r", marker="^", markersize=5, linestyle="dashed", linewidth=0.70)
plt.errorbar(tumor_volumes_pivot_mean.index, tumor_volumes_pivot_mean["Ketapril"], yerr=tumor_volumes_pivot_sem["Ketapril"], color="k", marker="s", markersize=5, linestyle="dashed", linewidth=0.70)
plt.errorbar(tumor_volumes_pivot_mean.index, tumor_volumes_pivot_mean["Placebo"], yerr=tumor_volumes_pivot_sem["Placebo"], color="g", marker="d", markersize=5, linestyle="dashed", linewidth=0.70)

plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'],loc="best", fontsize="medium", fancybox=True)

# Save the Figure
plt.savefig("results/Fig1.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_sites_mean = clinical_data_complete.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]

# Convert to DataFrame
met_sites_mean = pd.DataFrame(met_sites_mean)

# Preview DataFrame
met_sites_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sites_sem = clinical_data_complete.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
met_sites_sem = pd.DataFrame(met_sites_sem)

# Preview DataFrame
met_sites_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_sites_mean = met_sites_mean.reset_index()
met_sites_pivot_mean = met_sites_mean.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

met_sites_sem = met_sites_sem.reset_index()
met_sites_pivot_sem = met_sites_sem.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
met_sites_pivot_mean.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Capomulin"], yerr=met_sites_pivot_sem["Capomulin"], color="b", marker="o", markersize=5, linestyle="dashed", linewidth=0.70)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Infubinol"], yerr=met_sites_pivot_sem["Infubinol"], color="r", marker="^", markersize=5, linestyle="dashed", linewidth=0.70)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Ketapril"], yerr=met_sites_pivot_sem["Ketapril"], color="k", marker="s", markersize=5, linestyle="dashed", linewidth=0.70)
plt.errorbar(met_sites_pivot_mean.index, met_sites_pivot_mean["Placebo"], yerr=met_sites_pivot_sem["Placebo"], color="g", marker="d", markersize=5, linestyle="dashed", linewidth=0.70)

plt.title("Metastatic Spread During Treatment")
plt.ylabel("Met. Sites")
plt.xlabel("Treatment Duration (Days)")
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="medium", fancybox=True)

# Save the Figure
plt.savefig("results/Fig2.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = clinical_data_complete.groupby(["Drug", "Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame
mouse_count = pd.DataFrame({"Mouse Count": mouse_count})

# Preview DataFrame
mouse_count.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count = mouse_count.reset_index()
mouse_count_pivot = mouse_count.pivot(index="Timepoint", columns="Drug")["Mouse Count"]

# Preview the Data Frame
mouse_count_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(100 * mouse_count_pivot["Capomulin"] / 25, "bo", linestyle="dashed", markersize=5, linewidth=0.70)
plt.plot(100 * mouse_count_pivot["Infubinol"] / 25, "r^", linestyle="dashed", markersize=5, linewidth=0.70)
plt.plot(100 * mouse_count_pivot["Ketapril"] / 25, "ks", linestyle="dashed", markersize=5, linewidth=0.70)
plt.plot(100 * mouse_count_pivot["Placebo"] / 25 , "gd", linestyle="dashed", markersize=5, linewidth=0.70)

plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="medium", fancybox=True)

# Save the Figure
plt.savefig("results/Fig3.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_percent_change =  100 * (tumor_volumes_pivot_mean.iloc[-1] - tumor_volumes_pivot_mean.iloc[0]) / tumor_volumes_pivot_mean.iloc[0]
tumor_percent_change_sem =  100 * (tumor_volumes_pivot_sem.iloc[-1] - tumor_volumes_pivot_sem.iloc[0]) / tumor_volumes_pivot_sem.iloc[0]

# Display the data to confirm
tumor_percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_changes = (tumor_percent_change["Capomulin"], 
                   tumor_percent_change["Infubinol"], 
                   tumor_percent_change["Ketapril"], 
                   tumor_percent_change["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(percent_changes))  
barsPass = ax.bar(ind[0], percent_changes[0], 1, color='g')
barsFail = ax.bar(ind[1:], percent_changes[1:], 1, color='r')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-35,75])
ax.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(bars):
    for bar in bars:
        value = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., 5,
                '%d%%' % int(value),
                ha='center', va='bottom', color="k")

def autolabelPass(bars):
    for bar in bars:
        value = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., -10,
                '-%d%% ' % int(value),
                ha='center', va='bottom', color="k")

# Call functions to implement the function calls
autolabelPass(barsPass)
autolabelFail(barsFail)

# Save the Figure
fig.savefig("results/Fig4.png")

# Show the Figure
fig.show()
```
